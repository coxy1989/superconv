```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def final_mean_sd(result_dir):
    hxs = [pd.read_csv(f'{result_dir}/{i}.csv') for i in range(1,6)]
    val_accs = pd.DataFrame([hx['val_acc'] for hx in hxs]).T
    return  val_accs.mean(axis=1).iloc[-1], val_accs.std(axis=1).iloc[-1]

def plot_run(result_dir, label, ax):
    hxs = [pd.read_csv(f'{result_dir}/{i}.csv') for i in range(1,6)]
    val_accs = pd.DataFrame([hx['val_acc'] for hx in hxs]).T
    df = pd.DataFrame({'accuracy' : val_accs.mean(axis=1), 'epoch': list(range(1, val_accs.shape[0] + 1))})
    sns.lineplot(x='epoch', y='accuracy', data=df, label=label, ax=ax)
```

```python
mnist_exps = [['./experiment_mnist_1cycle_12', 'CLR(12)'],
              ['./experiment_mnist_1cycle_25', 'CLR(25)'],
              ['./experiment_mnist_1cycle_50', 'CLR(50)'],
              ['./experiment_mnist_1cycle_85', 'CLR(85)'],
              ['./experiment_mnist_inv', 'INV'],
              ['./experiment_mnist_reduce_on_plateau', 'ROP']
             ]

for file, _ in mnist_exps:
    print(final_mean_sd(file))


_, ax = plt.subplots(figsize=(700/192, 400/192), dpi=192)
for file, label in mnist_exps:
    plot_run(file, label, ax)
```

```python
cifar10_exps = [['./experiment_cifar10_1cycle_12', 'CLR(12)'],
                ['./experiment_cifar10_1cycle_15', 'CLR(15)'],
                ['./experiment_cifar10_1cycle_25', 'CLR(25)'],
                ['./experiment_cifar10_1cycle_50', 'CLR(50)'],
                ['./experiment_cifar10_1cycle_85', 'CLR(85)'],
                ['./experiment_cifar10_inv', 'INV'],
                ['./experiment_cifar10_reduce_on_plateau', 'ROP']]

for file, _ in cifar10_exps:
    print(final_mean_sd(file))

_, ax = plt.subplots(figsize=(700/192, 400/192), dpi=192)
for file, label in cifar10_exps:
    plot_run(file, label, ax)
```
