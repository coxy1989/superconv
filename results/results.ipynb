```python
import pandas as pd
import seaborn as sns
```

```python
def final_mean_sd(result_dir):
    hxs = [pd.read_csv(f'{result_dir}/{i}.csv') for i in range(1,6)]
    val_accs = pd.DataFrame([hx['val_acc'] for hx in hxs]).T
    return  val_accs.mean(axis=1).iloc[-1], val_accs.std(axis=1).iloc[-1]

def plot_run(result_dir):
    hxs = [pd.read_csv(f'{result_dir}/{i}.csv') for i in range(1,6)]
    val_accs = pd.DataFrame([hx['val_acc'] for hx in hxs]).T
    df = pd.DataFrame({'accuracy' : val_accs.mean(axis=1), 'epoch': list(range(1, val_accs.shape[0] + 1))})
    sns.lineplot(x='epoch', y='accuracy', data=df)
```

```python
mnist_exps = ['./experiment_mnist_1cycle_12',
              './experiment_mnist_1cycle_25',
              './experiment_mnist_1cycle_50',
              './experiment_mnist_1cycle_85',
              './experiment_mnist_inv',
              './experiment_mnist_reduce_on_plateau']

for e in mnist_exps:
    print(final_mean_sd(e))

for e in mnist_exps:
    plot_run(e)
```

```python
cifar10_exps = ['./experiment_cifar10_1cycle_12',
                './experiment_cifar10_1cycle_15',
                './experiment_cifar10_1cycle_25',
                './experiment_cifar10_1cycle_50',
                './experiment_cifar10_1cycle_85',
                './experiment_cifar10_inv',
                './experiment_cifar10_reduce_on_plateau'
               ]

for e in cifar10_exps:
    print(final_mean_sd(e))

for e in cifar10_exps:
    plot_run(e)
```
